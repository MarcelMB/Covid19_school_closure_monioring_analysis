```python
import pandas as pd 
import pymysql
from sqlalchemy import create_engine
import getpass  # To get the password without showing the input
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/unesco_school'
engine = create_engine(connection_string)
```

```python
#load csv
data_clean = pd.read_csv('cleaned_UNESCO_school_closures_database.csv')
```

```python
tableName   = 'data_clean' # name of your final dataframe
dataFrame   = pd.DataFrame(data = data_clean)
dbConnection    = engine.connect()
try:
    frame           = dataFrame.to_sql(tableName, dbConnection, if_exists='fail'); 
except ValueError as vx:
    print(vx)
except Exception as ex:
    print(ex)
else:
    print("Table %s created successfully."%tableName);
finally:
    dbConnection.close()
```
